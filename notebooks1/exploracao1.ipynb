### Lendo arquivo CSV

```python
path_csv = '../data_raw1/dados_empresaB.csv'
```

```python
with open(path_csv, 'r') as file:
    dados_csv = file.readlines()

dados_csv    
```

```python
# A identificação está vindo em string novamente, sem saber identificar linhas e colunas separadamente
dados_csv[1][0]
```

```python
#Importando a bliblioteca csv

import csv
```

```python
# Lendo os dados com o csv.reader separado por virgula (demiliter=)
# É necessário usar o obejto SPAMREADER. Ele é iterável, então deve ser percorrido com um FOR

dados_csv = [] # Lista vazia para receber os dados iterados pelo for

with open (path_csv, 'r') as file:
    spamreader = csv.reader(file, delimiter=',')
    for row in spamreader:
        dados_csv.append(row)
```

```python
dados_csv
```

```python
type(dados_csv[0])
```

```python
# Os dados do JSON estão em uma lista de dicionários, já os dados do CSV estão em uma lista de lista, e isso impossibilita a união dos dois arquivos
# Na lista de dicionarios temos o nome das colunas em cada um dos registros, já na lista de lista, apenas no primeiro registro tem o nome da coluna
```

```python
# No JSON o primeiro valor trás chaves e valores, ou seja, colunas e linhas
dados_json[0]
```

```python
# No CSV o primeiro valor está mostrando apenas os nomes das colunas
dados_csv[0]
```

```python
# Para essa situação, a lista de dicionarios se mostrou melhor para manipular os dados
```

```python
# Chamando a função DictReader para que os dados sejam lidos como lista de dicionarios. Ela foi incluída no lugar do csv.reader

dados_csv = []

with open (path_csv, 'r') as file:
    spamreader = csv.DictReader(file, delimiter=',')
    for row in spamreader:
        dados_csv.append(row)
```

```python
# Agora a váriavel "dados_csv" é uma lista de dicionários, assim como a váriável "dados_json"
dados_csv
```

```python
dados_csv[0]
```

# Transformando os dados
## Comparando campos dos arquivos json e csv (quantidade de linhas, colunas, etc)

```python
# Nomes das colunas são diferentes e o dados_csv possui uma coluna a mais (data da venda)
# Salvando os nomes das colunas em uma variável e transformando em uma lista

nome_colunas_json = list(dados_json[0].keys())
nome_colunas_json
```

```python
# Salvando os nomes das colunas em uma váriável e transformando em lista

nome_colunas_csv = list(dados_csv[0].keys())
nome_colunas_csv
```

```python
# Verificandoa a quantidade de colunas
len(nome_colunas_csv)
```

```python
# Verificandoa a quantidade de colunas
len(nome_colunas_json)
```

```python
# Renomeando os campos usando as colunas da empresaA (arquivo em json)
# Foi criado um novo dicionário para realizar a mudança dos nomes

key_mapping = {'Nome do Item': 'Nome do Produto',
                'Classificação do Produto': 'Categoria do Produto',
                'Valor em Reais (R$)': 'Preço do Produto (R$)',
                'Quantidade em Estoque': 'Quantidade em Estoque',
                'Nome da Loja': 'Filial',
                'Data da Venda': 'Data da Venda'}
key_mapping                
```

```python
# Atualizando os nomes
# Foi criado uma nova lista para receber esses dados transformados

new_dados_csv = []

for old_dict in dados_csv: # Aqui passaremos por cada um dos dados
    dict_temp = {}
    for old_key, value in old_dict.items(): # Aqui ele passará por cada uma das chaves, pegará o dado e criará uma nova coluna (chave e valor)
        dict_temp[key_mapping[old_key]] = value
    new_dados_csv.append(dict_temp)    

new_dados_csv[0]    
```

## União dos dados empresaA e empresaB

```python
# Verificando a quantidade de registros
len(dados_json)
```

```python
# Verificando a quantidade de registros
len(new_dados_csv)
```

```python
# Usando a função .extend() para criar a lista combinada

combined_list = []
combined_list.extend(dados_json)
combined_list.extend(new_dados_csv)
```

```python
len(combined_list)
```

```python
combined_list[0]
```

```python
combined_list[-1]
```

### Salvando os dados

```python
# Nova variável com os nomes de todas as colunas
nomes_colunas = list(combined_list[-1].keys())
nomes_colunas
```

```python
path_dados_combinados = '../data_processed1/dados_combinados.csv'

with open(path_dados_combinados, 'w') as file:
    writer = csv.DictWriter(file, fieldnames=nomes_colunas) # For para escrever um novo dicionário com os nomes da colunas
    writer.writeheader()

    for row in combined_list:
        writer.writerow(row)
```

```python
# Tratamento na coluna 'Data da venda' para os dados da empresaA, que não tem essa coluna

# nomes_colunas = list(combined_list[-1].keys())
# nomes_colunas
```

```python
# Metodo .get().A sintaxe básica é: dicionario.get(chave, valor_padrao)

dados_combinados_tabela = [nomes_colunas]

for row in combined_list:
    linha = []
    for coluna in nomes_colunas:
        linha.append(row.get(coluna, 'Indisponível'))

    dados_combinados_tabela.append(linha)  
```

```python
dados_combinados_tabela[1]  
```

### Salvando os dados

```python

path_dados_combinados = '../data_processed1/dados_combinados.csv'

with open(path_dados_combinados, 'w') as file:
    writer = csv.writer(file)
    writer.writerow(dados_combinados_tabela)
```
